```python
import pandas as pd

path = "/content/nyc_squirrels.csv"
df = pd.read_csv(path)
df
```

```python
print(df[df['unique_squirrel_id'] == "5D-PM-1018-05"])
```

```python
# Drop columns with more than 50% missing values (example threshold)
df.dropna(thresh=len(df) * 0.5, axis=1, inplace=True)
```

```python
df['highlight_fur_color'].fillna('Unknown', inplace=True)
df['primary_fur_color'].fillna('Unknown', inplace=True)
```

```python
#check for missing values

df.isnull().sum()
```

```python
df['primary_fur_color'] = df['primary_fur_color'].str.capitalize()
df['highlight_fur_color'] = df['highlight_fur_color'].str.capitalize()
```

```python
# Ensure hectare formatting (you can add custom logic here if you notice inconsistencies)
df['hectare'] = df['hectare'].str.upper()
```

```python
df['date'].head()
```

```python
# Convert the 'Date' column from MMDDYYYY to a datetime object
df['date'] = pd.to_datetime(df['date'], format='%m%d%Y')

# Preview the converted Date column
df.head()
```

```python
df.to_csv('/content/cleaned_nyc_squirrels.csv', index=False)
```
